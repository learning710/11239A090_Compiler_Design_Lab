```python
def leading_trailing(grammar):
    # Initialize LEADING and TRAILING sets for all non-terminals
    leading = {nt: set() for nt in grammar}
    trailing = {nt: set() for nt in grammar}

    def is_non_terminal(symbol):
        return symbol in grammar

    changed = True
    while changed:
        changed = False

        for nt, productions in grammar.items():
            for production in productions:

                # --- Calculate LEADING sets ---
                if not production or (len(production) == 1 and production[0] == ''):  # Production is epsilon: A -> ''
                    if '' not in leading[nt]:
                        leading[nt].add('')
                        changed = True
                else:
                    can_derive_epsilon_from_prefix = True
                    for symbol in production:
                        if is_non_terminal(symbol):
                            current_len = len(leading[nt])
                            leading[nt].update(s for s in leading[symbol] if s != '')
                            if len(leading[nt]) != current_len:
                                changed = True

                            if '' not in leading[symbol]:
                                can_derive_epsilon_from_prefix = False
                                break
                        else:  # symbol is a terminal
                            if symbol not in leading[nt]:
                                leading[nt].add(symbol)
                                changed = True
                            can_derive_epsilon_from_prefix = False
                            break # Stop considering subsequent symbols if a terminal is encountered

                    if can_derive_epsilon_from_prefix and '' not in leading[nt]:
                        leading[nt].add('')
                        changed = True


                # --- Calculate TRAILING sets ---
                if not production or (len(production) == 1 and production[0] == ''):  # Production is epsilon: A -> ''
                    if '' not in trailing[nt]:
                        trailing[nt].add('')
                        changed = True
                else:
                    can_derive_epsilon_from_suffix = True
                    for i in range(len(production) - 1, -1, -1):  # Iterate backwards
                        symbol = production[i]
                        if is_non_terminal(symbol):
                            current_len = len(trailing[nt])
                            trailing[nt].update(s for s in trailing[symbol] if s != '')
                            if len(trailing[nt]) != current_len:
                                changed = True

                            if '' not in trailing[symbol]:
                                can_derive_epsilon_from_suffix = False
                                break
                        else:  # symbol is a terminal
                            if symbol not in trailing[nt]:
                                trailing[nt].add(symbol)
                                changed = True
                            can_derive_epsilon_from_suffix = False
                            break # Stop considering preceding symbols if a terminal is encountered

                    if can_derive_epsilon_from_suffix and '' not in trailing[nt]:
                        trailing[nt].add('')
                        changed = True

    return leading, trailing
```

```python
grammar = {
    'S': [['A', 'B']],
    'A': [['a'], ['']],
    'B': [['b'], ['']]
}

print("Example Grammar:")
for nt, prods in grammar.items():
    for prod in prods:
        print(f"{nt} -> {' '.join(prod) if prod else 'ε'}")
```
