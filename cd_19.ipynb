```python
import ply.lex as lex
import ply.yacc as yacc

# ---------------- LEX ----------------
tokens = ('NUMBER',)

# Use literals for operators (simplest way)
literals = ['+', '-', '*', '/']

t_ignore = ' \t'

def t_NUMBER(t):
    r'\d+'
    t.value = int(t.value)
    return t

def t_error(t):
    print("Invalid character:", t.value[0])
    t.lexer.skip(1)

lexer = lex.lex()

# ---------------- YACC ----------------
stack = []

def p_expr_number(p):
    'expr : NUMBER'
    stack.append(p[1])

def p_expr_op(p):
    '''expr : expr expr '+'
            | expr expr '-'
            | expr expr '*'
            | expr expr '/' '''

    b = stack.pop()
    a = stack.pop()

    if p[3] == '+':
        stack.append(a + b)
    elif p[3] == '-':
        stack.append(a - b)
    elif p[3] == '*':
        stack.append(a * b)
    elif p[3] == '/':
        stack.append(a / b)

def p_error(p):
    print("Syntax error")

parser = yacc.yacc()

# ---------------- INPUT ----------------
expr = input("Enter postfix expression: ")

# Clear stack before each run
stack.clear()

parser.parse(expr)

# ---------------- OUTPUT ----------------
if len(stack) == 1:
    print("Result:", stack[0])
else:
    print("Invalid postfix expression")
```
