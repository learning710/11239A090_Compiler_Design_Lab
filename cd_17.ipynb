```python
import ply.lex as lex
import sys

# Workaround for AttributeError in Colab
if not hasattr(sys.modules['__main__'], '__file__'):
    sys.modules['__main__'].__file__ = 'colab_lexer_module_my.py'

class MyLexer:
    # List of token names
    tokens = ['ADD', 'ID', 'NUMBER']

    # Token regex rules
    t_ADD    = r'ADD'
    t_ID     = r'[a-zA-Z_][a-zA-Z0-9_]*'
    t_NUMBER = r'\d+'

    # Ignore spaces and tabs
    t_ignore = ' \t'

    # Handle newlines
    def t_newline(self, t):
        r'\n+'
        t.lexer.lineno += len(t.value)

    # Error handling rule (required!)
    def t_error(self, t):
        print(f"Illegal character '{t.value[0]}'")
        t.lexer.skip(1)

# Build the lexer by passing an instance of MyLexer and disabling optimization
lexer = lex.lex(object=MyLexer(), optimize=False, lextab=False)

# Test input
lexer.input("ADD item1 2")

# Print tokens
for tok in lexer:
    print(tok)
```
