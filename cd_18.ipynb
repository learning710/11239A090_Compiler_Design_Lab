```python
import ply.lex as lex
import types
import sys

# Define lexer rules and tokens globally first
tokens = [
    'VOWEL',
    'CONSONANT'
]

def t_VOWEL(t):
    r'[aeiouAEIOU]'
    return t

def t_CONSONANT(t):
    r'[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ]'
    return t

# A string containing ignored characters (spaces, tabs, newlines)
t_ignore = ' \t\n'

# Error handling rule
def t_error(t):
    print(f"Illegal character '{t.value[0]}' at position {t.lexpos}")
    t.lexer.skip(1)

# Create a dummy module object to store lexer components
# This is a common workaround for PLY in interactive environments like Colab/Jupyter
# where `ply` struggles with direct introspection of the '__main__' module.
lexer_module = types.ModuleType('my_vowel_consonant_lexer_module')
lexer_module.__file__ = 'my_vowel_consonant_lexer_module.py' # Provide a dummy file path

# Assign the lexer components to the dummy module
lexer_module.tokens = tokens
lexer_module.t_VOWEL = t_VOWEL
lexer_module.t_CONSONANT = t_CONSONANT
lexer_module.t_ignore = t_ignore
lexer_module.t_error = t_error

# Optionally, add the dummy module to sys.modules (good for introspection, though not strictly required for lex.lex)
sys.modules['my_vowel_consonant_lexer_module'] = lexer_module

# Build the lexer using the dummy module
lexer = lex.lex(module=lexer_module)

# --- Test the lexer ---
print("Testing lexer with 'Bakery':")
lexer.input("Bakery")
for tok in lexer:
    print(tok)

print("\nTesting lexer with 'apple':")
lexer.input("apple")
for tok in lexer:
    print(tok)

print("\nTesting lexer with 'Rhythm123':")
lexer.input("Rhythm123")
for tok in lexer:
    print(tok)
```
